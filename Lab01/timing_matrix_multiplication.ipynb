## Timing Matrix Multiplication
Here we compare the time it takes to multiply matrices using a standard method and a built in numpy method.

```python
from pylab import plot, title, xlabel, ylabel, show, legend
```

```python
from numpy import arange

# width of N x N matrices to be multiplied
N = arange(2, 301, 5)
```

```python
# initialize lists to hold calculation times
my_times = []
np_times = []
```

```python
from numpy import ones, zeros, dot
from time import clock

# run the calculations
for n in N:
    A = ones([n, n], float)*2
    B = ones([n, n], float)*3
    C = zeros([n, n], float)
    
    start = clock()
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i, j] += A[i, k]*B[k, j]
    end = clock()
    my_times.append(end-start)
    
    start = clock()
    C = dot(A, B)
    end = clock()
    np_times.append(end-start)
```

```python
from pylab import figure, plot, xlabel, ylabel, title, legend, show

figure(1)
plot(N, my_times, N, np_times)
xlabel('N'), ylabel('time [s]')
title('time vs matrix size for matrix multiplication')
legend(('my method', 'numpy.dot method'))
show()
```

```python
figure(2)
plot(N**3, my_times)
xlabel('N^3'), ylabel('time [s]')
title('my method times vs matrix sizes cubed')
show()
```
